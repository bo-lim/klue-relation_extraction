```python
from collections import Counter
import matplotlib.pyplot as plt
from train import label_to_num
from inference import num_to_label
import seaborn as sns
```

```python
from train import *
# load dataset
train_dataset = load_data("../dataset/train/train.csv")

train_label = label_to_num(train_dataset['label'].values)
```

```python
train_dataset
```

```python
c = Counter(train_dataset['label'])
sorted_c = sorted(c.items(),key=lambda x:x[1],reverse=True)
```

```python
c.items()
```

```python
labels = [i[0] for i in sorted_c[:10]]
print(num_to_label(labels))
```

```python
df_label = pd.DataFrame(sorted_c)
df_label.columns=['label','cnt']
```

```python

def grouping(label):
    sp = label.split(':')
    return sp[0]
def sep(labels):
    seps = []
    for label in labels:
        sp = label.split(':')
        seps.append(sp[-1])
    return seps
```

```python
df_label['group'] = df_label['label'].apply(grouping)
```

```python
sns.set_style('whitegrid')
sns.countplot(x='group',data=df_label)
plt.show()
```

```python
sns.barplot(x='label',y='cnt',hue='group',data=df_label)
plt.xticks(rotation=90)
plt.figure(figsize=(8,8))
plt.show()
```
